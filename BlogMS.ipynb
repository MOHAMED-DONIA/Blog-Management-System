## ⚙️ JavaScript — State & Helpers

```python
JS_HELPERS = '''
const API = 'http://localhost:8000';
let token = localStorage.getItem('blog_token') || null;
let currentUser = JSON.parse(localStorage.getItem('blog_user') || 'null');
let currentPage = 1, editingPostId = null, currentPostId = null;

function authHeaders() {
  return token
    ? { 'Authorization': `Bearer ${token}`, 'Content-Type': 'application/json' }
    : { 'Content-Type': 'application/json' };
}
async function apiFetch(path, opts = {}) {
  const res  = await fetch(API + path, { headers: authHeaders(), ...opts });
  const data = await res.json().catch(() => ({}));
  if (!res.ok) throw { status: res.status, detail: data.detail || 'Request failed' };
  return data;
}
function showAlert(id, msg, type = 'error') {
  const el = document.getElementById(id);
  el.textContent = msg;
  el.className = `alert alert-${type} show`;
  setTimeout(() => el.className = 'alert', 3500);
}
function roleBadge(role) { return `<span class="role-badge role-${role}">${role}</span>`; }
function formatDate(d)   { return new Date(d).toLocaleDateString('en-US', { year:'numeric', month:'short', day:'numeric' }); }
function escHtml(str)    { return String(str).replace(/&/g,'&amp;').replace(/</g,'&lt;').replace(/>/g,'&gt;').replace(/"/g,'&quot;'); }
function escAttr(str)    { return String(str).replace(/'/g,"\\'" ).replace(/\n/g,'\\n'); }
'''
print('JS_HELPERS defined ✅')
```

## ⚙️ JavaScript — Auth (Login / Register / Logout)

```python
JS_AUTH = '''
function showPanel(name) {
  document.querySelectorAll('.panel').forEach(p => p.classList.remove('active'));
  document.getElementById(name + '-panel').classList.add('active');
  if (name === 'posts') loadPosts(currentPage);
  if (name === 'admin') loadAdminUsers();
}
function switchAuthTab(tab) {
  document.querySelectorAll('.auth-tab').forEach((t,i) => t.classList.toggle('active',(i===0)===(tab==='login')));
  document.getElementById('login-form').style.display    = tab==='login'    ? '' : 'none';
  document.getElementById('register-form').style.display = tab==='register' ? '' : 'none';
}
async function doLogin() {
  const u = document.getElementById('login-username').value.trim();
  const p = document.getElementById('login-password').value;
  if (!u || !p) return showAlert('auth-alert','Please fill in all fields');
  try {
    const res  = await fetch(API+'/auth/login',{method:'POST',body:new URLSearchParams({username:u,password:p})});
    const data = await res.json();
    if (!res.ok) throw data;
    token = data.access_token;
    localStorage.setItem('blog_token', token);
    await fetchMe(); updateAuthUI(); showPanel('posts');
  } catch(e) { showAlert('auth-alert', e.detail||'Login failed'); }
}
async function doRegister() {
  const payload = {
    username: document.getElementById('reg-username').value.trim(),
    email:    document.getElementById('reg-email').value.trim(),
    password: document.getElementById('reg-password').value,
    role:     document.getElementById('reg-role').value,
  };
  if (!payload.username||!payload.email||!payload.password) return showAlert('auth-alert','Please fill in all fields');
  try {
    await apiFetch('/auth/register',{method:'POST',body:JSON.stringify(payload)});
    showAlert('auth-alert','Account created! Please sign in.','success');
    setTimeout(()=>switchAuthTab('login'),1200);
  } catch(e) { showAlert('auth-alert',e.detail||'Registration failed'); }
}
async function fetchMe() {
  try { currentUser=await apiFetch('/users/me'); localStorage.setItem('blog_user',JSON.stringify(currentUser)); }
  catch { logout(); }
}
function logout() {
  token=null; currentUser=null;
  localStorage.removeItem('blog_token'); localStorage.removeItem('blog_user');
  updateAuthUI(); showPanel('auth');
}
function updateAuthUI() {
  const loggedIn = !!token;
  document.getElementById('login-header-btn').style.display = loggedIn?'none':'';
  document.getElementById('logout-btn').style.display       = loggedIn?'':'none';
  const disp=document.getElementById('username-display');
  const newPostBtn=document.getElementById('new-post-btn');
  const adminBtn=document.getElementById('admin-panel-btn');
  if (loggedIn && currentUser) {
    disp.innerHTML = `${currentUser.username} ${roleBadge(currentUser.role)}`;
    newPostBtn.style.display = (currentUser.role==='admin'||currentUser.role==='author')?'':'none';
    adminBtn.style.display   = currentUser.role==='admin'?'':'none';
  } else { disp.textContent=''; newPostBtn.style.display='none'; adminBtn.style.display='none'; }
}
'''
print('JS_AUTH defined ✅')
```

## ⚙️ JavaScript — Admin Panel

```python
JS_ADMIN = '''
async function loadAdminUsers() {
  const container = document.getElementById('users-container');
  container.innerHTML = '<div class="loading"><span class="spinner"></span> Loading…</div>';
  try {
    const users = await apiFetch('/users/');
    container.innerHTML = `
      <table style="width:100%;border-collapse:collapse">
        <thead><tr style="border-bottom:1px solid var(--border);text-align:left">
          <th style="padding:0.75rem;color:var(--muted);font-size:0.8rem">ID</th>
          <th style="padding:0.75rem;color:var(--muted);font-size:0.8rem">Username</th>
          <th style="padding:0.75rem;color:var(--muted);font-size:0.8rem">Email</th>
          <th style="padding:0.75rem;color:var(--muted);font-size:0.8rem">Role</th>
          <th style="padding:0.75rem;color:var(--muted);font-size:0.8rem">Actions</th>
        </tr></thead>
        <tbody>${users.map(u => `
          <tr style="border-bottom:1px solid var(--border)">
            <td style="padding:0.75rem;font-family:var(--mono);font-size:0.8rem">#${u.id}</td>
            <td style="padding:0.75rem">${escHtml(u.username)}</td>
            <td style="padding:0.75rem;color:var(--muted)">${escHtml(u.email)}</td>
            <td style="padding:0.75rem">${roleBadge(u.role)}</td>
            <td style="padding:0.75rem">
              <select onchange="changeRole(${u.id},this.value)" ${u.username==='admin'?'disabled':''}
                style="background:var(--surface2);border:1px solid var(--border);color:var(--text);padding:0.3rem 0.5rem;border-radius:6px;font-size:0.8rem">
                <option value="reader"  ${u.role==='reader' ?'selected':''}>Reader</option>
                <option value="author"  ${u.role==='author' ?'selected':''}>Author</option>
                ${u.username==='admin'?'<option value="admin" selected>Admin</option>':''}
              </select>
              <button class="btn btn-danger btn-sm" style="margin-left:0.5rem"
                onclick="deleteUser(${u.id})" ${u.username==='admin'||u.id===currentUser.id?'disabled':''}>Delete</button>
            </td>
          </tr>`).join('')}
        </tbody>
      </table>`;
  } catch(e) { container.innerHTML='<div class="empty-state"><p>Failed to load users.</p></div>'; }
}
async function changeRole(userId, newRole) {
  try { await apiFetch(`/users/${userId}`,{method:'PUT',body:JSON.stringify({role:newRole})}); showAlert('admin-alert',`✅ Role updated to ${newRole}`,'success'); }
  catch(e) { showAlert('admin-alert',e.detail||'Failed to update role'); loadAdminUsers(); }
}
async function deleteUser(userId) {
  if (!confirm('Delete this user?')) return;
  try { await fetch(`${API}/users/${userId}`,{method:'DELETE',headers:authHeaders()}); loadAdminUsers(); }
  catch(e) { showAlert('admin-alert','Failed to delete user'); }
}
'''
print('JS_ADMIN defined ✅')
```

## ⚙️ JavaScript — Posts List & Pagination

```python
JS_POSTS = '''
async function loadPosts(page=1) {
  currentPage=page;
  const container=document.getElementById('posts-container');
  container.innerHTML='<div class="loading"><span class="spinner"></span> Loading…</div>';
  try {
    const data=await apiFetch(`/posts/?page=${page}&size=8`);
    document.getElementById('posts-meta').textContent=`${data.total} posts · page ${data.page}/${data.pages||1}`;
    container.innerHTML = data.items.length===0
      ? '<div class="empty-state"><p>No posts yet. Be the first to publish!</p></div>'
      : data.items.map(p=>`
          <div class="post-card" onclick="loadPostDetail(${p.id})">
            <div class="post-card-title">${escHtml(p.title)}</div>
            <div class="post-card-meta">
              <span>✍ Author #${p.author_id}</span>
              <span>📅 ${formatDate(p.created_at)}</span>
            </div>
          </div>`).join('');
    renderPagination(data.page, data.pages);
  } catch(e) { container.innerHTML='<div class="empty-state"><p>Failed to load posts.</p></div>'; }
}
function renderPagination(current, total) {
  const pg=document.getElementById('pagination');
  if (total<=1) { pg.innerHTML=''; return; }
  pg.innerHTML=`
    <button class="btn btn-ghost btn-sm" onclick="loadPosts(${current-1})" ${current<=1?'disabled':''}>← Prev</button>
    <span class="page-info">${current} / ${total}</span>
    <button class="btn btn-ghost btn-sm" onclick="loadPosts(${current+1})" ${current>=total?'disabled':''}>Next →</button>`;
}
'''
print('JS_POSTS defined ✅')
```

## ⚙️ JavaScript — Comments

```python
JS_COMMENTS = '''
async function loadComments(postId, page=1) {
  const cs=document.getElementById('comments-section');
  try {
    const data=await apiFetch(`/posts/${postId}/comments/?page=${page}&size=20`);
    cs.innerHTML=`
      <h2 class="comments-title">Comments (${data.total})</h2>
      ${!!token ? `
        <div class="comment-form">
          <div class="comment-form-title">Leave a comment</div>
          <textarea id="new-comment-text" placeholder="Write your comment…" style="min-height:80px;margin-bottom:0.75rem"></textarea>
          <button class="btn btn-primary btn-sm" onclick="submitComment(${postId},null)">Post Comment</button>
        </div>` : '<p style="color:var(--muted);font-size:0.85rem;margin-bottom:1.5rem">Sign in to leave a comment.</p>'}
      <div class="comment-list">${data.items.map(c=>renderComment(c,postId)).join('')}</div>`;
  } catch(e) { cs.innerHTML='<p style="color:var(--muted)">Failed to load comments.</p>'; }
}
function renderComment(c, postId) {
  const canDelete=currentUser&&(currentUser.role==='admin'||currentUser.id===c.author_id);
  const replies=(c.replies||[]).map(r=>renderComment(r,postId)).join('');
  return `
    <div class="comment-item" id="comment-${c.id}">
      <div class="comment-author">@${escHtml(c.author.username)} ${roleBadge(c.author.role)}</div>
      <div class="comment-content">${escHtml(c.content)}</div>
      <div class="comment-actions">
        ${!!token ? `<button class="btn btn-ghost btn-sm" onclick="toggleReply(${c.id})">↩ Reply</button>`:''}
        ${canDelete ? `<button class="btn btn-danger btn-sm" onclick="deleteComment(${postId},${c.id})">Delete</button>`:''}
      </div>
      <div class="reply-input-area" id="reply-area-${c.id}" style="display:none">
        <textarea placeholder="Write a reply…" id="reply-text-${c.id}" style="min-height:70px;margin-bottom:0.5rem"></textarea>
        <button class="btn btn-primary btn-sm" onclick="submitComment(${postId},${c.id})">Post Reply</button>
      </div>
    </div>${replies}`;
}
function toggleReply(commentId) {
  const area=document.getElementById(`reply-area-${commentId}`);
  area.style.display=area.style.display==='block'?'none':'block';
}
async function submitComment(postId, parentId) {
  const textEl=parentId?document.getElementById(`reply-text-${parentId}`):document.getElementById('new-comment-text');
  const content=textEl.value.trim();
  if (!content) return;
  try { await apiFetch(`/posts/${postId}/comments/`,{method:'POST',body:JSON.stringify({content,parent_id:parentId||null})}); textEl.value=''; loadComments(postId); }
  catch(e) { showAlert('detail-alert',e.detail||'Failed to post comment'); }
}
async function deleteComment(postId, commentId) {
  if (!confirm('Delete this comment?')) return;
  try { await fetch(`${API}/posts/${postId}/comments/${commentId}`,{method:'DELETE',headers:authHeaders()}); loadComments(postId); }
  catch(e) { showAlert('detail-alert','Failed to delete comment'); }
}
'''
print('JS_COMMENTS defined ✅')
```

## ⚙️ JavaScript — Post CRUD (Create / Edit / Delete)

```python
JS_POST_CRUD = '''
function openModal() {
  editingPostId=null;
  document.getElementById('modal-title').textContent='New Post';
  document.getElementById('modal-post-title').value='';
  document.getElementById('modal-post-content').value='';
  document.getElementById('post-modal').classList.add('show');
}
function editPost(id, title, content) {
  editingPostId=id;
  document.getElementById('modal-title').textContent='Edit Post';
  document.getElementById('modal-post-title').value=title;
  document.getElementById('modal-post-content').value=content;
  document.getElementById('post-modal').classList.add('show');
}
function closeModal() { document.getElementById('post-modal').classList.remove('show'); }
async function submitPost() {
  const title  =document.getElementById('modal-post-title').value.trim();
  const content=document.getElementById('modal-post-content').value.trim();
  if (!title||!content) return showAlert('modal-alert','Title and content are required');
  try {
    if (editingPostId) {
      await apiFetch(`/posts/${editingPostId}`,{method:'PUT',body:JSON.stringify({title,content})});
      closeModal(); loadPostDetail(editingPostId);
    } else {
      const post=await apiFetch('/posts/',{method:'POST',body:JSON.stringify({title,content})});
      closeModal(); loadPostDetail(post.id);
    }
  } catch(e) { showAlert('modal-alert',e.detail||'Failed to save post'); }
}
async function deletePost(postId) {
  if (!confirm('Delete this post permanently?')) return;
  try { await fetch(`${API}/posts/${postId}`,{method:'DELETE',headers:authHeaders()}); showPanel('posts'); }
  catch(e) { showAlert('detail-alert','Failed to delete post'); }
}
'''
print('JS_POST_CRUD defined ✅')
```
